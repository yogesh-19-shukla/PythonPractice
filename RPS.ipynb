```python
import random
```

```python
def getChoices():
  player_choice = input("Enter a choice: rock / paper / scissors: ")
  options = ["rock","paper","scissors"]
  computer_choice = random.choice(options)
  choices = {"player":player_choice, "computer":computer_choice}
  return choices

def checkWinner(player, computer):
  print(f"You chose: {player}, computer chose: {computer}")
  if player == computer:
    return "It's a tie!"
  elif player == "rock":
    if computer == "paper":
      return "Computer Wins"
    else:
      return "You Win"
  elif player == "paper":
    if computer == "rock":
      return "You Win"
    else:
      return "Computer Wins"
  elif player == "scissors":
    if computer == "rock":
      return "Computer Wins"
    else:
      return "You Win"

choices = getChoices()
winner = checkWinner(choices["player"], choices["computer"])       ## takes choices from dict key
print(winner)
```
